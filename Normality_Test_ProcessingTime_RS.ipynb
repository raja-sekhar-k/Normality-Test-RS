## Normality Test
## The processing time of purchase orders are given. Is the processing time normally destributed?

```python
import pandas as pa
from scipy import stats as mystats
import matplotlib.pyplot as myplot
```

```python
myData = pa.read_csv('.\datasets\PO_Processing.csv')
```

```python
myData
```

```python
PT = myData.Processing_Time
```

```python
PT
```

```python
mystats.probplot(PT, plot = myplot)
myplot.show()
```

```python
mystats.mstats.normaltest(PT)
```

```python
# P Value is more than 0.05 (P>0.05) , So the distribution seems to be normal
```
